```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='view', y='price', data=kc_house_data)
plt.title('Влияние вида на стоимость')
plt.xlabel('Уровень вида')
plt.ylabel('Цена (в долларах)')
plt.grid(axis='y')
plt.show()
```

Выводы:
  - Объекты с более высоким уровнем вида имеют обычно более высокую цену.
  - Однако есть некоторое количество выбросов.

```python
# Создадим два поднабора данных: с видом на воду и без вида на воду
waterfront_yes = kc_house_data[kc_house_data['waterfront'] == 1]
waterfront_no = kc_house_data[kc_house_data['waterfront'] == 0]

# Посчитаем количество домов с видом на воду и без вида на воду
count_waterfront_yes = len(waterfront_yes)
count_waterfront_no = len(waterfront_no)

# Посчитаем средние цены для каждой категории
avg_price_waterfront_yes = waterfront_yes['price'].mean()
avg_price_waterfront_no = waterfront_no['price'].mean()

# Создадим данные для круговой диаграммы
labels = ['С водным видом', 'Без водного вида']
sizes = [count_waterfront_yes, count_waterfront_no]

# Строим круговую диаграмму
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Соотношение домов с водным видом и без водного вида')
plt.axis('equal')  # Сделаем круговую диаграмму круглой
plt.show()

# Выведем средние цены для каждой категории
print("Средняя цена для домов с видом на воду:", avg_price_waterfront_yes)
print("Средняя цена для домов без вида на воду:", avg_price_waterfront_no)
```
